```python
# basic
import pandas as pd
import talib


#time
# import datetime as datetime

#manual
import crypto
from backtesting import Backtest , Strategy
from backtesting.lib import SignalStrategy , TrailingStrategy , crossover
```

```python
def SMA(values, n):
    """
    Return simple moving average of `values`, at
    each step taking into account `n` previous values.
    """
    
    return talib.SMA(values,n)
#     return pd.Series(values).rolling(n).mean()

def BIAS(values, n):
    
    if n == 1:
#         a1 = talib.SMA(values,n)
        a2 = values/values -1
    else:
        a1 = talib.SMA(values,n)
        a2 = (values/a1 -1)  *100

    return a2

class SmaCross(Strategy):
    # Define the two MA lags as *class variables*
    # for later optimization
    n1 = 5
    n2 = 65
    
    def init(self):
        # Precompute the two moving averages
        self.sma1 = self.I(SMA, self.data.Close, self.n1)
        self.sma2 = self.I(SMA, self.data.Close, self.n2)
    
    def next(self):
        # If sma1 crosses above sma2, close any existing
        # short trades, and buy the asset
        if crossover(self.sma1, self.sma2):
            self.position.close()
            self.buy()

        # Else, if sma1 crosses below sma2, close any existing
        # long trades, and sell the asset
        elif crossover(self.sma2, self.sma1):
            self.position.close()
            self.sell()

class BiasOver(Strategy):
    # Define the two MA lags as *class variables*
    # for later optimization
    n1 = 1
    n2 = 4
    n3 = 50
    
    a1 = 2
    def init(self):
        super().init()
        # Precompute the two moving averages
        self.bias1 = self.I(BIAS, self.data.Close, self.n1)
        self.bias2 = self.I(BIAS, self.data.Close, self.n2)
        self.ema1 = self.I(talib.EMA, self.data.Close, self.n3)

    def next(self):
        super().next()
  

#          if self.ema1 < self.data.Close:
#             # If sma1 crosses above sma2, close any existing
#             # short trades, and buy the asset
#             if (self.bias2-self.bias1 < self.a1*-1) & (self.bias2[-2]-self.bias1[-2] >self.a1*-1) :
#                 self.position.close()
#                 self.buy()
#         else:
#             # Else, if sma1 crosses below sma2, close any existing
#             # long trades, and sell the asset
#             if (self.bias2-self.bias1 > self.a1) & (self.bias2[-2]-self.bias1[-2] <self.a1):
#                 self.position.close()
#                 self.sell()

        # If sma1 crosses above sma2, close any existing
        # short trades, and buy the asset
        if (self.bias2-self.bias1 < self.a1*-1) & (self.bias2[-2]-self.bias1[-2] >self.a1*-1) :
            self.position.close()
            self.buy()

        # Else, if sma1 crosses below sma2, close any existing
        # long trades, and sell the asset
        elif (self.bias2-self.bias1 > self.a1) & (self.bias2[-2]-self.bias1[-2] <self.a1):
            self.position.close()
            self.sell()

```
